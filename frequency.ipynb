```python
#mylong = newlong
#mylat = newlat
def xydict(mylong, mylat):
    p = []
    thres = 0.025
    upx = int(0.5 / thres)
    upy = int(0.3 / thres)

    time = len(mylong)
    datasize = len(mylong[0])
    for i in range(datasize):
        p.append(dict())
        for j in range(time):
            x = int ((mylong[j][i]-113.8) / thres)
            y = int ((mylat[j][i]-22.5) / thres)
            if 0<=x<=upx and 0<=y<=upy:
                if (x,y) in p[i]:
                    p[i][(x,y)] += 1
                else:
                    p[i][(x,y)] = 1
    return p
```

```python
def ccw(A, B, C):
    """Tests whether the turn formed by A, B, and C is ccw"""
    return (B.x - A.x) * (C.y - A.y) > (B.y - A.y) * (C.x - A.x)

def intersect(a1, b1, a2, b2):
    """Returns True if line segments a1b1 and a2b2 intersect."""
    return ccw(a1, b1, a2) != ccw(a1, b1, b2) and ccw(a2, b2, a1) != ccw(a2, b2, b1)

```
